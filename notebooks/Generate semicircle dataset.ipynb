```python
%matplotlib notebook
import sys
import math
import numpy as np
import os.path as osp
from bunch import Bunch
sys.path.append(osp.dirname(osp.dirname(osp.abspath('__file__'))))

import utils.data_manager as dm
import acoc.acoc_plotter as ap
```

```python
def generate_sm_dataset(size=1000):
    el_per_class = int(size / 2)
    a = dm.semi_circle_gaussian(size=el_per_class)
    c_range = dm.MinMax(0, math.pi)
    b = dm.semi_circle_gaussian(size=el_per_class, c_range=c_range, center=(1.8, -.7))
    data = np.concatenate((a, b), axis=0)
    target = np.concatenate((np.full(el_per_class, 0, dtype=int), np.full(el_per_class, 1, dtype=int)), axis=0)
    return Bunch({'data': data, 'target': target})
```

```python
dt = generate_sm_dataset(1000)
d = np.concatenate((dt.data, np.array([dt.target]).T), axis=1)
```

```python
ap.plot_data(d)
```
